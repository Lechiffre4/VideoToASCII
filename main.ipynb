```python
import cv2
import ascii_magic
import imgkit
import os
```

```python
#Get Base Video
cap = cv2.VideoCapture("./WorkingFile/BaseVid.mp4")
#Get Base Video FPS
fps = cap.get(cv2.CAP_PROP_FPS)
#Get Base Video Width
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
#Get Base Video Height
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
#Get Base Video Length
length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

#extract sound from video with opencv

```

```python
#Get Frames from Base Video
if not os.path.exists('./WorkingFile/FramesFromBaseVid'):
    os.mkdir('./WorkingFile/FramesFromBaseVid')
else:
    for i in os.listdir('./WorkingFile/FramesFromBaseVid'):
        os.remove('./WorkingFile/FramesFromBaseVid/' + i)
        
count = 0
flag = 1
while flag:
    flag, image = cap.read()
    try:
        cv2.imwrite('./WorkingFile/FramesFromBaseVid/BaseFrame%d.jpg' % count, image)
    except:
        break
    count += 1
```

```python
#Convert Frames to HTML Files
if not os.path.exists('./WorkingFile/HTML_Frames'):
    os.mkdir('./WorkingFile/HTML_Frames')
else:
    for file in os.listdir('./WorkingFile/HTML_Frames'):
        os.remove('./WorkingFile/HTML_Frames/' + file)

for i in range(count):
    HTMLFrame = ascii_magic.AsciiArt.from_image('./WorkingFile/FramesFromBaseVid/BaseFrame%d.jpg' % i)
    #Resize HTML Frames
    HTMLFrame.to_html_file('./WorkingFile/HTML_Frames/HTML_Frame%d.html' % i, columns=200)
```

```python
#Convert HTML Files to jpg
if not os.path.exists('./WorkingFile/AsciiFrames'):
    os.mkdir('./WorkingFile/AsciiFrames')
else:
    for file in os.listdir('./WorkingFile/AsciiFrames'):
        os.remove('./WorkingFile/AsciiFrames/' + file)

config = imgkit.config(wkhtmltoimage='./wkhtmltopdf/bin/wkhtmltoimage.exe')

for i in range(count):
    imgkit.from_file('./WorkingFile/HTML_Frames/HTML_Frame%d.html' % i, './WorkingFile/AsciiFrames/AsciiFrame%d.jpg' % i, config=config)
```

```python
#Convert jpg to mp4
os.mkdir('./WorkingFile/AsciiVideo')

frame = cv2.imread('./WorkingFile/AsciiFrames/AsciiFrame0.jpg')
height, width, layers = frame.shape
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
video = cv2.VideoWriter('./WorkingFile/AsciiVideo/AsciiVid.mp4', fourcc, fps, (width,height))
for i in range(count):
    video.write(cv2.imread('./WorkingFile/AsciiFrames/AsciiFrame%d.jpg' % i))
cv2.destroyAllWindows()
video.release()
```
